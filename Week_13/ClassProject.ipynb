```python
import pandas as pd

# Create a dictionary with the company data
data = {
    'Company': ['Enron', 'Anderson', 'GK Jones', 'Mica', 'Dune'],
    'Founded': [1987, 1938, 2001, 1996, 2008],
    'Shares': [1000000, 1500000, 3000000, 250000, 800000],
    'Liabilities': [200000, 1500000, 50000, 300000, 200000]
}

# Create a DataFrame from the dictionary
df = pd.DataFrame(data)

# Calculate the assets column by subtracting liabilities from shares
assets = df['Shares'] - df['Liabilities']

# Calculate the over-leverage column by dividing (Assets - Liabilities) by Assets and multiplying by 100
df['Assets'] = assets
df['Over-Leverage (%)'] = ((df['Assets'] - df['Liabilities']) / df['Assets']) * 100

# Save the DataFrame to an Excel file using openpyxl as the engine
df.to_excel('Companny_data.xlsx', index=False, engine='openpyxl')
```
